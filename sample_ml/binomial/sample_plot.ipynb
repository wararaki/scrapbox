```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb
from scipy.stats import binom

%matplotlib inline
```

```python
N = 100
p = 0.5
y = np.array([i+1 for i in range(N)])
```

```python
def bino(N, y, p):
    return comb(N, y) * np.power(p, y) * np.power((1-p), (N-y))
```

```python
P = bino(N, y, p)
```

```python
plt.plot(P)
```

```python
P_sp = binom.pmf(y, N, p)
```

```python
plt.plot(P_sp)
```
