```python
import numpy as np
import math
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0.01, 6, 10000)
```

x > 0 制約

```python
def gamma(x):
    f = lambda t: np.exp(-t)*np.power(t, x-1)
    return quad(f, 0, np.inf)
```

```python
gamma_x = np.vectorize(gamma)(x)
```

```python
plt.plot(x, gamma_x[0])
```

math.gamma を使う

```python
gamma_x = np.vectorize(math.gamma)(x)
```

```python
plt.plot(x, gamma_x)
```
