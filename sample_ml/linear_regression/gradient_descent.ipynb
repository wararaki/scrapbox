```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.array([1, 1, 2, 3, 4, 5, 5, 6, 7, 8]).reshape(-1, 1)
y = np.array([1, 2, 2, 4, 4, 5, 6, 6, 6, 9]).reshape(-1, 1)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
def gradient_descent(x, y, n_iter=100, eta=0.001):
    w = np.random.rand(x.shape[1], 1)
    print(w)
    m = x.shape[0]
    for _ in range(n_iter):
        gradients = 2/m * x.T.dot(x.dot(w) - y)
        w = w - eta * gradients
    return w
```

```python
w = gradient_descent(x, y)
```

```python
x_pred = np.arange(x.shape[0])
y_pred = w*x_pred
```

```python
plt.scatter(x, y)
plt.plot(x_pred, y_pred[0])
plt.show()
```
