```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.array([1, 1, 2, 3, 4, 5, 5, 6, 7, 8])
y = np.array([1, 2, 2, 4, 4, 5, 6, 6, 6, 9])
```

```python
plt.scatter(x, y)
plt.show()
```

```python
def least_square(x, y):
    mu_x = x.mean()
    mu_y = y.mean()
    a = np.sum((x - mu_x)*(y - mu_y)) / np.sum(np.power(x-mu_x, 2))
    b = mu_y - a * mu_x
    return a, b
```

```python
def f(x, a, b):
    return a*x + b
```

```python
a, b = least_square(x, y)
x_pred = np.arange(x.shape[0])
y_pred = f(x_pred, a, b)
```

```python
plt.scatter(x, y)
plt.plot(x_pred, y_pred)
plt.show()
```
