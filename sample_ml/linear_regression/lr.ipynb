```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_classification

%matplotlib inline
```

```python
def lr(x, w):
    return w * x;
```

```python
X, label = make_classification(n_samples=100)
y = X[:, 1]
X = X[:, 0]
```

```python
w = np.array([0.5 for i in range(X.shape[0])]).reshape(1, -1)
b = 0
```

```python
y_pred = lr(X, w)[0]
```

```python
fig, ax = plt.subplots()
plt.scatter(X, y, c=label)
plt.plot(X, y_pred)
```
