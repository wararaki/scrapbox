```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_classification

%matplotlib inline
```

```python
X, label = make_classification(n_samples=100)
y = X[:, 1]
X = X[:, 0]
```

```python
w = np.array([0.5 for i in range(X.shape[0])]).reshape(1, -1)
b = 0
```

```python
def lr(x, w, b):
    return w * x + b;
```

```python
y_pred = lr(X, w, b)[0]
```

init

```python
fig, ax = plt.subplots()
plt.scatter(X, y, c=label)
plt.plot(X, y_pred)
```

```python
def fit(x, y_true, w, b, n=10):
    for i in range(n):
        y_pred = lr(x, w, b)[0]
        w = np.sum(y_pred - np.mean(y_true))
```
