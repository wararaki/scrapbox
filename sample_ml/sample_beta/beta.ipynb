```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad

%matplotlib inline
```

```python
x = np.linspace(0, 1, num=100)
```

```python
a = 2
b = 3

def beta(x):
    return (x**(a-1))*((1-x)**(b-1))

def f(x):
    C = 1 / quad(beta, 0, 1)[0]
    return C * (x**(a-1)) * ((1-x)**(b-1))
```

```python
y = f(x)
```

```python
plt.plot(y)
```
