```python
import matplotlib.pyplot as plt
```

```python
data = [15, 30, 10, 45]
```

```python
plt.pie(data, startangle=90)
circle = plt.Circle((0, 0), 0.5, color='white')
plt.axis('equal')
p = plt.gcf()
p.gca().add_artist(circle)
```

```python
plt.pie(data, startangle=90, wedgeprops=dict(width=0.5))
plt.axis('equal')
```
