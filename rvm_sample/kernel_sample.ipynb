```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def kernel(x_i, x_j, beta=1.):
    return np.exp(-beta * np.sum((x_i - x_j)**2))
```

```python
def plot_graph(X, Y):
    plt.figure(figsize=(7, 4))
    plt.plot(X, Y)
    plt.show()
```

```python
# create basedata
def get_init_data():
    X = np.arange(-1.5, 1.5, 0.01)
    Y = np.zeros(len(X))
    return X, Y
```

きれいな分布

```python
X, Y = get_init_data()
for i, x in enumerate(X):
    Y[i] = kernel(x, 0, 10)
plot_graph(X, Y)
```

適当にずらしてみる。

```python
centers = [-1, 0, 0.5, 0.6, 1]
weights = [0.2, 1, 0.3, -1, 0.5]
```

中心をずらしてみる。

```python
X, Y = get_init_data()
for i, x in enumerate(X):
    for center in centers:
        Y[i] += kernel(X[i], center, 10)
plot_graph(X, Y)
```

中心＋重さをずらしてみる。

```python
X, Y = get_init_data()
for i, x in enumerate(X):
    for weight, center in zip(weights, centers):
        Y[i] += weight * kernel(X[i], center, 10)
plot_graph(X, Y)
```
