```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import datasets

%matplotlib inline
```

```python
iris = datasets.load_iris()
data = np.hstack((iris.data, iris.target.reshape(-1, 1)))
df = pd.DataFrame(data, columns=iris.feature_names+['label'])
```

```python
df.head(3)
```

```python
ax = sns.boxplot(x='label', y='sepal length (cm)', data=df)
ax = sns.stripplot(x='label', y='sepal length (cm)', data=df)
```

```python
ax = sns.boxplot(x='label', y='sepal width (cm)', data=df)
ax = sns.stripplot(x='label', y='sepal width (cm)', data=df)
```

```python
ax = sns.boxplot(x='label', y='petal length (cm)', data=df)
ax = sns.stripplot(x='label', y='petal length (cm)', data=df)
```

```python
ax = sns.boxplot(x='label', y='petal width (cm)', data=df)
ax = sns.stripplot(x='label', y='petal width (cm)', data=df)
```
