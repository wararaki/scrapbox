```python
import itertools
import numpy as np
import pandas as pd
from scipy import sparse
```

```python
n_item = 20
n_user = 10
user_item_matrix = np.random.randint(2, size=n_user*n_item).reshape((n_user, -1))
user_item_matrix = sparse.lil_matrix(user_item_matrix)
```

```python
pd.DataFrame(user_item_matrix.todense(), index=[f'user_{i}' for i in range(n_user)], columns=[f'item_{i}' for i in range(n_item)])
```

```python
user_item_dict = {}
for i, row in enumerate(user_item_matrix.rows):
    user_item_dict[i] = set(row)
```

```python
def dice(set_1, set_2):
    total = len(set_1)+len(set_2)
    if total == 0:
        return 0
    return 2* len(set_1.intersection(set_2)) / total
```

```python
dice_similarities = sparse.lil_matrix((n_user, n_user))
for i, j in itertools.combinations(np.arange(n_user), 2):
    dice_similarities[i, j] = dice(user_item_dict[i], user_item_dict[j])
    dice_similarities[j, i] = dice_similarities[i, j]
for i in range(n_user):
    dice_similarities[i, i] = 1
dice_similarites = dice_similarities.tocsc()
```

```python
pd.DataFrame(dice_similarities.todense(), index=[f'user_{i}' for i in range(n_user)], columns=[f'user_{i}' for i in range(n_user)])
```
