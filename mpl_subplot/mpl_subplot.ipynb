```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris.keys()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head(3)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
df.plot.scatter(x='sepal length (cm)', y='sepal width (cm)', alpha=0.5, ax=axs[0, 0])
sns.regplot(x='sepal length (cm)', y='sepal width (cm)', data=df, ax=axs[0, 1])
df.plot.hist(x='sepal length (cm)', y='sepal width (cm)', bins=10, ax=axs[1, 0])
sns.distplot(df['sepal length (cm)'], bins=10, ax=axs[1, 1])
```
